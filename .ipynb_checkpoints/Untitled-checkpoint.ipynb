```python
from tensorflow.keras.layers import Rescaling , Dense , MaxPool2D , Conv2D , Input , Flatten
```

```python
model = Sequential([
  layers.Rescaling(1./255, input_shape=(60,60, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(2)
])
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'] )
```

```python
hist = model.fit(train_ds , epochs = 40 , validation_split = 0.2 , batch_size = 32)
```
